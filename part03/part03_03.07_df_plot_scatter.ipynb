### - 산점도

UCI 자동차 연비 데이터셋을 이용하여 두 변수의 관계를 나타내는 산점도를 그린다.<br>
`plot()` 메소드에 `kind='scatter'` 옵션을 넣고, 데이터프레임의 열 중에서 서로 비교할 두 변수를 선택한다.<br>
x축에 차량의 무게 데이터를 갖는 'weight' 열을 지정하고, y축에는 연비를 나타내는 'mpg' 열을 지정한다.
<br><br>
산점도를 보면 차량의 무게가 클수록 연비는 전반적으로 낮아지는 경향을 보인다.<br>
차량의 무게와 연비는 역(-)의 상관관계를 갖는다고 해석할 수 있다.

```python
import pandas as pd

# read_csv() 함수로 df 생성
df = pd.read_csv('./auto-mpg.csv', header=None)

# 열 이름 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name']

# 2개의 열을 선택하여 산점도 그리기
df.plot(x='weight', y='mpg', kind='scatter')
```
